```python
import ifcopenshell
from ifcopenshell.geom import create_shape
from ifcopenshell import util

# Load the IFC model
#ifc_model = ifcopenshell.open('path/to/your/file.ifc')

# Set up geometry settings
settings = ifcopenshell.geom.settings()

# Function to get the location of an object (position and orientation)
def get_location(entity):
    # Check if the entity has a Local Placement
    if entity.is_a('IfcProduct'):
        placement = entity.ObjectPlacement
        if placement and placement.is_a('IfcLocalPlacement'):
            # Extract the location (position) and orientation (rotation)
            placement_location = placement.RelativePlacement.Location
            return placement_location
    return None

# Function to extract geometry and location for each product (object)
for entity in ifc_model.by_type('IfcProduct'):
    try:
        # Get the geometry for the product (e.g., wall, door, etc.)
        shape = create_shape(settings, entity)
        geometry = shape.geometry
        
        # Get the location of the entity
        location = get_location(entity)

        # Print geometry and location information (for debugging or processing)
        print(f"Entity: {entity}")
        print(f"Geometry: {geometry}")  # This will be the raw geometry data (can be further processed)
        
        if location:
            print(f"Location: {location}")
        else:
            print("No location found.")
        
        print("-" * 50)  # Separator for readability
    except Exception as e:
        print(f"Error processing {entity}: {e}")
```

```python
# Retrieve the geometry (triangulated mesh) and location for an entity
for entity in ifc_model.by_type('IfcProduct'):
    try:
        # Create shape for the entity
        shape = create_shape(settings, entity)
        geometry = shape.geometry

        # Get location (position)
        location = get_location(entity)
        if location:
            position = location.Coordinates
            print(f"Entity: {entity.GlobalId} Location: {position}")
        
    except Exception as e:
        print(f"Error processing {entity}: {e}")
```

```python
import ifcopenshell

# Load the IFC file
ifc_file = ifcopenshell.open('DummyModel.ifc')

# Find the IfcColumn entity by its unique identifier or type
column = ifc_file.by_type('IfcColumn')[0]

# Traverse related entities to get the IfcFacetedBrep and IfcClosedShell
shape = column.Representation.Representations[0]
brep = shape.Items[0]
shell = brep.Outer

# Extract coordinates from the IfcClosedShell
points = set()
for face in shell.CfsFaces:
    for bound in face.Bounds:
        loop = bound.Bound
        for vertex in loop.Polygon:
            points.add(vertex.Coordinates)

# Find min and max for each axis
min_x = min(p[0] for p in points)
max_x = max(p[0] for p in points)
min_y = min(p[1] for p in points)
max_y = max(p[1] for p in points)
min_z = min(p[2] for p in points)
max_z = max(p[2] for p in points)

# Calculate dimensions
width = max_x - min_x
depth = max_y - min_y
height = max_z - min_z

print(points)
print(f'Width: {width}, Depth: {depth}, Height: {height}')
```
